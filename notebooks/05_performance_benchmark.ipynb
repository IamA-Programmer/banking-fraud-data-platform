```python
import time

df = spark.read.format("delta").load("/delta/silver/transactions_with_customers")

start = time.time()

df.groupBy("user_id").count().show()

end = time.time()

print("time without optimization ", end - start)

# Optimization
spark.conf.set("spark.sql.shuffle.partitions" , 100)

spark_optimized = df.repartition("user_id")

start = time.time()

spark_optimized.groupBy("user_id").count().show()

end = time.time()

print("Time with optimization ", end - start)


```
